```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
#stopwords.words('english')
```

```python
port_stem = PorterStemmer()
import re
```

```python
def stemming(content):
    stemmed_content = re.sub('[^a-zA-Z]',' ',content)
    stemmed_content = stemmed_content.lower()
    stemmed_content = stemmed_content.split()
    stemmed_content = [ port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
    stemmed_content = ' '.join(stemmed_content)
    return stemmed_content
```

```python
product_descriptions1 = product_descriptions1.apply(stemming)
```

### Creating TFIDF vectorization

TF=No of times a term occurs in document/ total number of terms in the document


IDF= log(Total number of doc)/number of documents with term t in it

```python
# (doc_no, index of that word) , log thing
# cardinality (len of a set)
```

```python
#What were actually doing is that we are marking our unique words 
#with anindex number and that index number is visible in a set on a second position in a set displayed 
#so now whenever there is written a zero or one the zero in first position in set that tells us  document number
#we are using for example in this Corpus we have 71 unique words and averi is marked as 4 by sckit learn.
```
